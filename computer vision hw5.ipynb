```python
### homework 5
'''
gradient magnitude images and choose proper thresholds to get the binary edge images (可參考 HW9)
(a) Roberts operator： performs a simple, quick to compute, 2-D spatial gradient measurement on an image            
(b) Prewitt edge detector：
(c) Sobel edge detector：
'''
import cv2
import numpy as np

image=cv2.imread('lena.bmp',0)

def Robertsoperator(img):
    kernal_G1=([[-1,0],[0,1]])
    kernal_G2=([[0,-1],[1,0]])
    row,column=img.shape
    temp_img=cv2.copyMakeBorder(src=img,top=0,bottom=1,left=0,right=1,borderType=cv2.BORDER_REPLICATE)
    newImg=img.copy().astype(float)
    for i in range(row):
        for j in range(column):
            temp=temp_img[i:i+2,j:j+2]
            newImg[i,j]=np.sqrt(np.sum(np.multiply(kernal_G1,temp))**2+np.sum(np.multiply(kernal_G2,temp))**2)
            #new_img[i,j] = np.sqrt(np.sum(np.multiply(ker_r1, temp))**2 + np.sum(np.multiply(ker_r2, temp))**2)
            #print(np.multiply(temp,kernal_G2))
            #print(np.multiply(kernal_G2,temp)) 上和下兩個相同並非是矩陣乘法，而是在兩個陣陣在相同位置的element相乘     
    return newImg    
    
def Prewittedgeoperator(img):
    kernal_G1=([[-1,-1,-1],[0,0,0],[1,1,1]])
    kernal_G2=([[-1,0,1],[-1,0,1],[-1,0,1]])
    row,column=img.shape
    temp_img=cv2.copyMakeBorder(src=img,top=1,bottom=1,left=1,right=1,borderType=cv2.BORDER_REPLICATE)
    newImg=img.copy().astype(float)
    for i in range(row):
        for j in range(column):
            temp=temp_img[i:i+3,j:j+3]
            newImg[i,j]=np.sqrt(np.sum(np.multiply(kernal_G1,temp))**2+np.sum(np.multiply(kernal_G2,temp))**2)
    return newImg
    
def Sobeledgeoperator(img):
    kernal_G1=([[-1,-2,-1], [0,0,0], [1,2,1]])
    kernal_G2=([[-1,0,1], [-2,0,2], [-1,0,1]])
    row,column=img.shape
    temp_img=cv2.copyMakeBorder(src=img,top=1,bottom=1,left=1,right=1,borderType=cv2.BORDER_REPLICATE)
    newImg=img.copy().astype(float)
    for i in range(row):
        for j in range(column):
            temp=temp_img[i:i+3,j:j+3]
            newImg[i,j]=np.sqrt(np.sum(np.multiply(kernal_G1,temp))**2+np.sum(np.multiply(kernal_G2,temp))**2)
    return newImg

'''
def ReverseThreshold(img,threshold):
    newImg=np.empty(img.shape)
    newImg.fill(255)
    '
    一般的binary pic
    如果計算得到的灰階值大於或等於該門檻值，則將此點的顏色設為白色(1)；
    如果計算得到的灰階值是小於該門檻值，則將此點的顏色設為黑色(0)。
    '
    mask=img>=threshold
    print(mask)
    print('after the thresold')
    print(newImg[mask])
    newImg[mask]=0
    return newImg

'''

def reverse_thresholding(img, threshold=128):
    new_img = np.empty(img.shape)
    new_img.fill(255)
    mask = img >= threshold
    new_img[mask] = 0
    return  new_img


Roberts_Img=Robertsoperator(image)
#type(Roberts_Img) -->numpy.ndarray
#plt.imshow(Roberts_Img)
Prewitt_Img=Prewittedgeoperator(image)
Sobel_Img=Sobeledgeoperator(image)

cv2.imwrite('Roberts.bmp',Roberts_Img)
cv2.imwrite('Prewitt.bmp',Prewitt_Img)
cv2.imwrite('Sobel.bmp',Sobel_Img)


cv2.imwrite('Robersoperator.bmp',reverse_thresholding(Roberts_Img,30)) 
#應該把黑白的部分反過來，讓邊線是黑的其他呈白色
cv2.imwrite('Prewittoperator.bmp',reverse_thresholding(Prewitt_Img,90)) 
cv2.imwrite('Sobeloperator.bmp',reverse_thresholding(Sobel_Img,130)) 



```
