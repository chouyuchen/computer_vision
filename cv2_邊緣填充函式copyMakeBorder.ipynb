```python
## 

import cv2
import numpy as np
import matplotlib.pyplot as plt

image=cv2.imread('lena.bmp',0)
plt.figure(figsize=(5,5))
plt.imshow(image,interpolation='bicubic')

img2=cv2.copyMakeBorder(image,200,200,200,200,cv2.BORDER_REPLICATE)
plt.figure(figsize=(5,5))
plt.imshow(img2,interpolation='bicubic')

img3 = cv2.copyMakeBorder(image,200,200,200,200,cv2.BORDER_REFLECT)
plt.figure(figsize=(5,5))
plt.imshow(img3,interpolation='bicubic')

img4 = cv2.copyMakeBorder(image,200,200,200,200,cv2.BORDER_REFLECT_101)
plt.figure(figsize=(5,5))
plt.imshow(img4,interpolation='bicubic')

img5 = cv2.copyMakeBorder(image,200,200,200,200,cv2.BORDER_WRAP)
plt.figure(figsize=(5,5))
plt.imshow(img5,interpolation='bicubic')

img6 = cv2.copyMakeBorder(image,10,10,10,10,cv2.BORDER_CONSTANT)
plt.figure(figsize=(5,5))
plt.imshow(img6,interpolation='bicubic')

#
RED = [255,0,0]
img7 = cv2.copyMakeBorder(image,10,10,10,10,cv2.BORDER_CONSTANT,value=RED)
plt.figure(figsize=(5,5))
plt.imshow(img7)


# https://www.jianshu.com/p/055706fd32ee


# https://www.itread01.com/content/1545194709.html
```
